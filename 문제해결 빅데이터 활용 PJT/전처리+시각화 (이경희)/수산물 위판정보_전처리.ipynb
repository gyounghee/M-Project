````python
def value_avg(df) :
  for row in range(len(df)-1):
    if ( df.loc[row,'위판일자'] == df.loc[row+1,'위판일자'] ) and ( df.loc[row,'산지조합명'] == df.loc[row+1,'산지조합명'] ) and ( df.loc[row,'수산물표준코드명'] == df.loc[row+1,'수산물표준코드명'] ) :
      df.loc[row+1,'평균가'] = ( (df.loc[row,'평균가']) + df.loc[row+1,'평균가'] ) / 2
      df.drop( [row], axis=0, inplace = True)  
````
