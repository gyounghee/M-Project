```python
# 데이터 불러오기
apt = pd.read_csv('/content/drive/MyDrive/멀티캠퍼스/data/데이터 시각화PJT/아파트_매매_실거래_평균가격(https:  kosis.kr statHtml statHtml.do?orgId=408&tblId=DT_KAB_11672_S15&conn_path=I2).csv', encoding='cp949')

# 컬럼명 변경
apt['행정구역'] = ['전국','서울특별시','부산광역시','대구광역시','인천광역시','광주광역시','대전광역시','울산광역시','세종특별자치시','경기도','강원도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주특별자치도']

# 세종시 '-'를 데이터 0으로 초기화
for col in apt.columns[1:]:  
  apt[col].replace('-',0, inplace=True)

# ------------ 년도별 평균 구하기 ----------
apt_avg = deepcopy(apt) 

# 컬럼 타입 변경
for col in apt_avg.columns[1:] :
  apt_avg = apt_avg.astype({col:'float'})

# 2020년까지 각 연도별 평균 컬럼 생성 및 값 초기화
for i in range(15):         
  apt_avg[f'{i+2006}'] = 0.0   
  for j in range(len(apt_avg)):          
    apt_avg[f'{i+2006}'][j] = round(sum(apt_avg.iloc[j, (i*12)+1 : ((i+1)*12)+1 ]) /12, 2) # 각 연도 평균 (round함수 이용해서 소수점 제한)
# 2020년 전까지 월별 데이터 삭제
for col in apt.columns[1:] :   
  if col == '2021. 01': break    
  del apt_avg[col]            

# 2021년도 평균 컬럼 생성 및 값 초기화
apt_avg['2021'] = 0.0    
for n in range(len(apt_avg)):      
  apt_avg['2021'][n] = round( sum( apt_avg.iloc[n, 1:11] )/10 ,2) 
# 2021년도 월별 데이터 삭제
for col in apt_avg.columns[1:11] :
  del apt_avg[col]  
```
