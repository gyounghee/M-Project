## 데이터 불러오기

```python
gender_age_homeless = pd.read_csv('/content/drive/MyDrive/멀티캠퍼스/[D9&10] 데이터시각화 프로젝트_4조/Data/인구 데이터/거주지역_가구주의_성_연령대별_무주택_가구수_20152020(httpskosis.krstatHtmlstatHtml.doorgId=101&tblId=DT_1OH0418&conn_path=I3).csv', encoding='cp949')
gender_age_homeless.head(3)
```

## 컬럼명 변경 및 불필요한 데이터 삭제

- 컬럼명 변경  
`연도`    →  `총계`  
`연도.1`  →  `30세미만`  
`연도.2`  →  `30~39세`  
`연도.3`  →  `40~49세`  
`연도.4`  →  `50~59세`  
`연도.5`  →  `60~69세`  
`연도.6`  →  `70~79세`  
`연도.7`  →  `80세이상`  

```python
n = 1  # 컬럼 위치를 나타내기 위한 변수 (2열부터 시작하려고 1으로 초기화 함)
for Y in range(2015,2021):    # 연도
  for _ in range(8) :  #  각 연도마다 8개의 카테고리  
    n+=1  # 다음 컬럼을 보기 위해 
    if gender_age_homeless.columns[n] == f'{Y}':    gender_age_homeless.rename(columns={f'{Y}':f'{Y}(총계)'}, inplace=True)       
    if gender_age_homeless.columns[n] == f'{Y}.1':  gender_age_homeless.rename(columns={f'{Y}.1':f'{Y}(30세미만)'}, inplace=True)  
    if gender_age_homeless.columns[n] == f'{Y}.2':  gender_age_homeless.rename(columns={f'{Y}.2':f'{Y}(30~39세)'}, inplace=True)  
    if gender_age_homeless.columns[n] == f'{Y}.3':  gender_age_homeless.rename(columns={f'{Y}.3':f'{Y}(40~49세)'}, inplace=True) 
    if gender_age_homeless.columns[n] == f'{Y}.4':  gender_age_homeless.rename(columns={f'{Y}.4':f'{Y}(50~59세)'}, inplace=True) 
    if gender_age_homeless.columns[n] == f'{Y}.5':  gender_age_homeless.rename(columns={f'{Y}.5':f'{Y}(60~69세)'}, inplace=True) 
    if gender_age_homeless.columns[n] == f'{Y}.6':  gender_age_homeless.rename(columns={f'{Y}.6':f'{Y}(70~79세)'}, inplace=True) 
    if gender_age_homeless.columns[n] == f'{Y}.7':  gender_age_homeless.rename(columns={f'{Y}.7':f'{Y}(80세이상)'}, inplace=True) 
gender_age_homeless.head(1)
```

- 컬럼명 변경   
`거주지역별(1)` → `거주지역`

```python
gender_age_homeless.rename(columns={'거주지역별(1)':'행정구역','성별(1)':'성별'}, inplace=True)
```

- 0행 삭제 & 인데스 재정의

```python
gender_age_homeless.drop(index=[0],inplace=True)
gender_age_homeless.reset_index(drop=True, inplace=True)
```

## 수도권, 지방 추가

```python
# 수도권    - 3,4,5, 12,13,14, 27,28,29
gender_age_homeless.loc[54] = gender_age_homeless.loc[0]   # 총계
gender_age_homeless.loc[55] = gender_age_homeless.loc[1]   # 남자
gender_age_homeless.loc[56] = gender_age_homeless.loc[2]   # 여자
gender_age_homeless.loc[54, '행정구역'] = '수도권'
gender_age_homeless.loc[55, '행정구역'] = '수도권'
gender_age_homeless.loc[56, '행정구역'] = '수도권'

# 지방 -  0,1,2,3,4,5,12,13,14,27,28,29 를 제외한 모든 값
gender_age_homeless.loc[57] = gender_age_homeless.loc[0]   # 총계
gender_age_homeless.loc[58] = gender_age_homeless.loc[1]   # 남자
gender_age_homeless.loc[59] = gender_age_homeless.loc[2]   # 여자
gender_age_homeless.loc[57, '행정구역'] = '지방'
gender_age_homeless.loc[58, '행정구역'] = '지방'
gender_age_homeless.loc[59, '행정구역'] = '지방'


# 각 해당 카테고리에 해당하는 인덱스 설정
지방_총계 = [6,9]
지방_총계 += [i for i in range(15,54,3)]
지방_남자 = [7,10]
지방_남자 += [i for i in range(16,54,3)]
지방_여자 = [8,11]
지방_여자 += [i for i in range(17,54,3)]

# 수도권, 지방 컬럼값 재정의
for i in range(2, len(gender_age_homeless.columns)):
  # 수도권
  gender_age_homeless.iloc[54, i] =int(sum( gender_age_homeless.iloc[[3, 12, 27], i].astype(int) )/3)  # 총계
  gender_age_homeless.iloc[55, i] =int(sum( gender_age_homeless.iloc[[4, 13, 28], i].astype(int) )/3)  # 남자
  gender_age_homeless.iloc[56, i] =int(sum( gender_age_homeless.iloc[[5, 14, 29], i].astype(int) )/3)  # 여자
  # 지방
  gender_age_homeless.iloc[57, i] =int(sum( gender_age_homeless.iloc[[1, 4, 9], i].astype(int) )/3) # 총계
  gender_age_homeless.iloc[58, i] =int(sum( gender_age_homeless.iloc[[1, 4, 9], i].astype(int) )/3) # 남자
  gender_age_homeless.iloc[59, i] =int(sum( gender_age_homeless.iloc[[1, 4, 9], i].astype(int) )/3) # 여자

# 인데스 재정의
idx = [0,1,2,54,55,56,57,58,59]
add = [i for i in range(3,54)]

gender_age_homeless = gender_age_homeless.reindex(index= idx + add )
gender_age_homeless.reset_index(drop=True, inplace = True) 
```
