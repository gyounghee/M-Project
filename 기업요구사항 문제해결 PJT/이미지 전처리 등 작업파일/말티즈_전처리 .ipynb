```python
import os, sys, json, cv2
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

import shutil
```

## set_1(이경희) 적용

```python
#### 모든 이미지 선명하게  - set_1
def sharp(name, file_path, save_path):
  file = os.listdir(file_path)

  i = 0
  for img_name in file:
    image = cv2.imread(f"{file_path}/{img_name}") 
    kernel = np.array([[0, -1, 0],
                    [-1, 5,-1],
                    [0, -1, 0]]) # 커널 생성
    # 이미지를 선명하게 만듦
    image_sharp = cv2.filter2D(image, -1, kernel)
    # save
    i += 1
    cv2.imwrite(f'{save_path}/sharp_{name}_{i}.jpg',image_sharp)
    
## 정상
name = 'nor'
file_path = '/content/drive/MyDrive/멀티캠퍼스/3조/Dog_Pic/최종수집자료/합계/말티즈/정상'
save_path = '/content/drive/MyDrive/멀티캠퍼스/3조/이경희/말티즈/set_1/말티즈_정상'
sharp(name,file_path, save_path)

## 비만
name = 'fat'
file_path = '/content/drive/MyDrive/멀티캠퍼스/3조/Dog_Pic/최종수집자료/합계/말티즈/비만'
save_path = '/content/drive/MyDrive/멀티캠퍼스/3조/이경희/말티즈/set_1/말티즈_비만'
sharp(name,file_path, save_path)
```

## set_5(이경희) 적용


```python
#### 모든 이미지에 적용 전 테스트
image = cv2.imread(f"/content/drive/MyDrive/멀티캠퍼스/3조/Dog_Pic/최종수집자료/합계/말티즈/정상/1000.jpg") 
kernel = np.array([[0, -1, 0],
                [-1, 5,-1],
                [0, -1, 0]]) # 커널 생성
# 이미지를 선명하게 만듦
image_sharp = cv2.filter2D(image, -1, kernel)
image_yuv = cv2.cvtColor(image_sharp, cv2.COLOR_BGR2YUV) # YUV로 변경
image_yuv[:, :, 0] = cv2.equalizeHist(image_yuv[:, :, 0]) # 히스토그램 평활화 적용
plt.imshow(image_yuv), plt.axis("off") # 이미지 출력
plt.show()
```

```python
#### 모든 이미지에 적용
def square_sharp_contrast(name, file_path, save_path):
  file = os.listdir(file_path)

  i = 0
  for img_name in file:
    image = cv2.imread(f"{file_path}/{img_name}") 
    kernel = np.array([[0, -1, 0],
                    [-1, 5,-1],
                    [0, -1, 0]]) # 커널 생성
    # 이미지를 선명하게 만듦
    image_sharp = cv2.filter2D(image, -1, kernel)
    image_yuv = cv2.cvtColor(image_sharp, cv2.COLOR_BGR2YUV) # YUV로 변경
    image_yuv[:, :, 0] = cv2.equalizeHist(image_yuv[:, :, 0]) # 히스토그램 평활화 적용
    # save
    i+=1
    cv2.imwrite(f'{save_path}/square_sharp_contrast_{i}.jpg',image_yuv)

## 정상
name = 'nor'
file_path = '/content/drive/MyDrive/멀티캠퍼스/3조/Dog_Pic/최종수집자료/합계/말티즈/정상'
save_path = '/content/drive/MyDrive/멀티캠퍼스/3조/이경희/말티즈/set_5/말티즈_정상'
square_sharp_contrast(name,file_path, save_path)

## 비만
name = 'fat'
file_path = '/content/drive/MyDrive/멀티캠퍼스/3조/Dog_Pic/최종수집자료/합계/말티즈/비만'
save_path = '/content/drive/MyDrive/멀티캠퍼스/3조/이경희/말티즈/set_5/말티즈_비만'
square_sharp_contrast(name,file_path, save_path)
```

```python
def edge(name, file_path, save_path):
  file = os.listdir(file_path)
    
  i=0
  for img_name in file:
      image = cv2.imread(f'{file_path}/{img_name}')
      sobelx = cv2.Sobel(image, cv2.CV_8U, 1, 0, ksize=3)

      ## save
      i+=1
      cv2.imwrite(f'{save_path}/SolX_{name}_{i}.jpg',sobelx)

## 정상
name = 'nor'
file_path = '/content/drive/MyDrive/멀티캠퍼스/3조/Dog_Pic/최종수집자료/합계/말티즈/정상'
save_path = '/content/drive/MyDrive/멀티캠퍼스/3조/이경희/말티즈/set_4/말티즈_정상'
edge(name,file_path, save_path)

## 비만
name = 'fat'
file_path = '/content/drive/MyDrive/멀티캠퍼스/3조/Dog_Pic/최종수집자료/합계/말티즈/비만'
save_path = '/content/drive/MyDrive/멀티캠퍼스/3조/이경희/말티즈/set_4/말티즈_비만'
edge(name,file_path, save_path)
```
