```python
def stringToRGB(base64_string):
    imgdata = base64.b64decode(base64_string)
    dataBytesIO = io.BytesIO(imgdata)
    image = Image.open(dataBytesIO)
    return cv2.cvtColor(np.array(image), cv2.COLOR_BGR2RGB)

######### 바로 아래 코드처럼 이미지처럼 바꾸지 않고 np.array 형태로 저장해야함
##### img = Image.fromarray( stringToRGB(base64_str) )
# cv2.imwrite('/content/drive/MyDrive/멀티캠퍼스/3조/이경희/sss.jpg',stringToRGB(base64_str))
# stringToRGB(base64_str)
```
