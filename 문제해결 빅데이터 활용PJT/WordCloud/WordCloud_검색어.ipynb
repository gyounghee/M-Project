```python
import pickle

word_list = []
for col in range(len(df.columns)):   # 총 19열
  for row in range(len(df)):  # 총 61행
    if int(df.iloc[row,col]) == 0 : break
    else : 
      for n in range(int(df.iloc[row,col])):
        word_list.append( df.columns[col]) 

filePath = '/content/drive/MyDrive/멀티캠퍼스/2조/이경희/WordCloud/wordcloud.txt'

# txt파일로 쓰기
with open(filePath, 'wb') as lf:
  pickle.dump(word_list,lf)

# txt파일 읽기
with open(filePath, 'rb') as lf:
  words = pickle.load(lf)

# dict 생성
word_list = {}
for w in words:
  try :
    if word_list[w] >= 1: word_list[w] += 1
  except :
    word_list[w] = 1
```

```python
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import matplotlib.pyplot as plt
import pickle
import pandas as pd
import numpy as np

# mask
icon = Image.open('/content/drive/MyDrive/멀티캠퍼스/2조/이경희/WordCloud/한반도.png')
mask = Image.new("RGB", icon.size, (255,255,255))
mask.paste(icon, icon)
mask = np.array(mask)

# wordcloud
filePath = '/content/drive/MyDrive/멀티캠퍼스/2조/이경희/WordCloud/wordcloud.txt'
font_path = '/content/drive/MyDrive/멀티캠퍼스/2조/이경희/WordCloud/ttf/BMDOHYEON_ttf.ttf'

wc = WordCloud(font_path = font_path, mask = mask, colormap='Blues', background_color='white')
cloud = wc.generate_from_frequencies(word_list)

# show
plt.figure(figsize=(10, 15))
plt.axis('off')
plt.imshow(wc)


# wc.to_file('/content/drive/MyDrive/멀티캠퍼스/2조/이경희/WordCloud/검색어_wordcloud6.png')
```
